```python
import pandas as pd
from IPython.display import display
import os 

df = pd.read_csv('Hpatch_summary.csv')
```

```python
df_all = df.sort_values('Time Avg')

base_cols = ['Matcher', 'Time Avg']
```

```python
def highlight_best_3(s, ascending=False):
    sorted_vals = sorted(s.unique(), reverse=not ascending)
    
    val_1 = sorted_vals[0] if len(sorted_vals) > 0 else None
    val_2 = sorted_vals[1] if len(sorted_vals) > 1 else None
    val_3 = sorted_vals[2] if len(sorted_vals) > 2 else None
    
    styles = []
    for val in s:
        if val == val_1:
            styles.append('background-color: #abebc6; color: black; font-weight: bold') 
        elif val == val_2:
            styles.append('background-color: #aed6f1; color: black') 
        elif val == val_3:
            styles.append('background-color: #f9e79f; color: black') 
        else:
            styles.append('')
    return styles
```

```python
print("Table 1: TOTAL Metrics")

total_cols = [c for c in df.columns if c.startswith('Total')]
final_cols = base_cols + total_cols

tabel_full = df_all[final_cols].style \
    .format(precision=2) \
    .hide(axis='index') \
    .apply(highlight_best_3, subset=total_cols, ascending=False) \
    .apply(highlight_best_3, subset=['Time Avg'], ascending=True)

display(tabel_full)
```

```python
print("|Table 2: ILLUMINATION Metrics")

illum_cols = [c for c in df.columns if c.startswith('Illumination')]
cols_table2 = base_cols + illum_cols

style_illum = df_all[cols_table2].style \
    .format(precision=2) \
    .hide(axis='index') \
    .apply(highlight_best_3, subset=illum_cols, ascending=False) \
    .apply(highlight_best_3, subset=['Time Avg'], ascending=True)

display(style_illum)
```

```python
print("Table 3: VIEWPOINT Metrics")

view_cols = [c for c in df.columns if c.startswith('Viewpoint')]
cols_table3 = base_cols + view_cols

style_view = df_all[cols_table3].style \
    .format(precision=2) \
    .hide(axis='index') \
    .apply(highlight_best_3, subset=view_cols, ascending=False) \
    .apply(highlight_best_3, subset=['Time Avg'], ascending=True)

display(style_view)
```

```python
def generate_table_html(dataframe, columns, title):
    top3 = {}
    for col in [c for c in columns if c != 'Matcher']:
        is_asc = (col == 'Time Avg')
        top3[col] = sorted(dataframe[col].unique(), reverse=not is_asc)[:3]

    colors = {0: '#abebc6', 1: '#aed6f1', 2: '#f9e79f'} # Hijau, Biru, Kuning

    # Header HTML
    html = [f'<h3>{title}</h3><table border="1" style="border-collapse: collapse; width: 100%; font-family: Arial; margin-bottom: 20px;">']
    html.append('<thead><tr style="background-color: #f2f2f2;">' + 
                ''.join([f'<th style="padding: 8px; border: 1px solid #ddd;">{c}</th>' for c in columns]) + 
                '</tr></thead><tbody>')

    for _, row in dataframe[columns].iterrows():
        row_html = []
        for col in columns:
            val = row[col]
            style = 'padding: 8px; text-align: center; border: 1px solid #ddd;'
            
            # Cek Warna
            if col in top3 and val in top3[col]:
                rank = top3[col].index(val)
                if rank < 3:
                    style += f' background-color: {colors[rank]}; color: black;' + (' font-weight: bold;' if rank==0 else '')
            
            display_val = f"{val:.2f}" if isinstance(val, float) else str(val)
            row_html.append(f'<td style="{style}">{display_val}</td>')
        html.append(f'<tr>{"".join(row_html)}</tr>')
    
    html.append('</tbody></table>')
    return "".join(html)

full_html = ['<h1>HPatches Evaluation Report</h1>']

full_html.append(generate_table_html(df_all, final_cols, "Table 1: TOTAL Metrics"))
full_html.append(generate_table_html(df_all, cols_table2, "Table 2: ILLUMINATION Metrics"))
full_html.append(generate_table_html(df_all, cols_table3, "Table 3: VIEWPOINT Metrics"))

full_html.append('<br><div><strong>Color Legend:</strong> <span style="background-color:#abebc6; border:1px solid #ccc; padding:2px 10px;">1st Best</span> <span style="background-color:#aed6f1; border:1px solid #ccc; padding:2px 10px;">2nd Best</span> <span style="background-color:#f9e79f; border:1px solid #ccc; padding:2px 10px;">3rd Best</span></div>')

output_path = os.path.join('..', 'docs', 'HPATCH_REPORT.md')
os.makedirs(os.path.dirname(output_path), exist_ok=True)

with open(output_path, 'w') as f:
    f.write("\n".join(full_html))

print(f"{os.path.abspath(output_path)} generated successfully.")
```
